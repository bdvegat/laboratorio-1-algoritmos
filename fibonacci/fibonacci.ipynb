Fibonacci

```python
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
from datetime import timedelta

def fibonacci(n):
    n1=0
    n2=1
    fib=0
    if n==0:
        return 0
    if n==1:
        return 1
    for i in range (1,n):
        fib=n1+n2
        n1=n2
        n2=fib
    return fib
```

```python
n=[]
exTime=[]
for i in range (0,79999,1000):
    t1=datetime.now()
    p=fibonacci(i)
    t2=datetime.now()
    n.append(i)
    exTime.append((t2-t1).total_seconds())
```

```python
plt.plot(n,exTime)
plt.xlabel('n')
plt.ylabel('tiempo de ejecucion');
```
